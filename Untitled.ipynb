```python
import pandas as pd
```

```python
# 37.51041488512401, 127.0382883094826
# 위도 0.00001˚ 차이는 약 1.1m 거리에 해당하며, 경도 0.00001˚ 차이는 약 0.9m 거리에 해당
gn = [0,0]
```

```python
d = 0.4
h = (d**2-(d/2)**2)**0.5
d,h
```

```python
def make_subs(positon):
    
    sub_1 = [positon[0]+d, positon[1]]
    sub_2 = [positon[0]+(d/2), positon[1]+h]
    sub_3 = [positon[0]-(d/2), positon[1]+h]
    sub_4 = [positon[0]-d, positon[1]]
    sub_5 = [positon[0]-(d/2), positon[1]-h]
    sub_6 = [positon[0]+(d/2), positon[1]-h]
    return [sub_1,sub_2,sub_3,sub_4,sub_5,sub_6]
```

```python
make_subs(gn)
```

```python
all_positions = [gn]+make_subs(gn)
all_positions
```

```python
all_positions_set = {}

i = 1
while len(all_positions_set) < 200:
    all_positions += make_subs(all_positions[i])
    all_positions_tuple = [tuple(l) for l in all_positions]
    all_positions_set = set(all_positions_tuple)
#     all_positions = list(all_positions_set)
    i += 1
```
